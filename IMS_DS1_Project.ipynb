# Information about attributes

---



**Pelvis Incidence :**
Sagital Allignment of spine and pelvis.   

**Pelvis Tilt :**
Angle showing the Allignment of spine and pelvis.   

**Lumber Lordoisis Angle :**
A curving inward of the lower back is called Lumber Lordoisis. In our data, measurement of its angle is recorded of different patients.

**Sacral Slope :**
The sacral slope is measured between the tangent line to the superior endplate of and the horizontal plane. Sacral inclination is defined as the angle between the vertical plane and the tangential line to the sacral dorsum.

**Pelvis Radius :**
Radius of the sagital allignment which is used as an technique to relate with Lumber Lordoisis.

**Spondylolisthesis Degree :**
Spondylolisthesis is a spinal disorder in which a bone (vertebra) slips forward onto the bone below it. Degree of spondylolisthesis represents the influence of spondylolisthesis in an individual patient.

**Pelvis Slope :**
Slope of the sagital allignment.

**Direct Tilt:**
Tilt position of sagital allignment.

**Thoracic Slope :**
The thoracic inlet, also known as the superior thoracic aperture, refers to the opening at the top of the thoracic cavity. In our data, we have its slope measurement of individual patients.

**Cervical Tilt :**
In anatomy, cervical is an adjective that is related to **Neck**. We have its tilt measurement.

**Sacrum Angle :**
Sacrum is a triangular bone in the lower back formed from fused vertebrae and situated between the two hip bones of the pelvis. We have its angle measurement of patients.

**Scoliosis Slope**
Scoliosis is a medical condition in which a person's spine has a sideways curve.We have its slope measurement.

# K-Means Clustering Technique

---

**k-means clustering** is a popular method for cluster analysis in data mining. k-means clustering aims to partition n observations into k clusters in which each observation belongs to the cluster with the nearest mean, serving as a prototype of the cluster.
![K-means](https://upload.wikimedia.org/wikipedia/commons/d/d5/Kmeans_animation.gif)

Major aim is to find **optimal distances from the centroid to individual data points**(depends on problem context) which satisfies that the datapoints are grouped correctly. (**Note :** k-means is computationaly **hard** algorithm, the partitions that contains objets of similar type may have some members that should belong to other group)

![k-means-1](https://dashee87.github.io/images/kmeans.gif)

Statiscal Evaluation are done using **ANOVA**(**AN**ALYSIS **O**F **VA**RIANCE) method because :

> to indicates that the **significance levels** should not be looked at as the **test of equal means**, as the **cluster solution** has been derived based on Euclidean distance to minimize the distance.



> for **determining value of k** if we use **elbow method**.





For doing cluster analysis using k-means, we require **`sklearn`** package. K-means is present in **`cluster`** sub-package.

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=2)
kmeans.fit(train)
```

```python
kmeans.labels_
```

```python
prediction = kmeans.predict(test)
correct = sum(prediction)
print("Acurracy of model is ",(correct/93)*100,"%")
```

```python
kmeans1 = KMeans(n_clusters=2, init='k-means++',n_init=10,max_iter=50, algorithm="elkan")
kmeans1.fit(train)
```

```python
prediction = kmeans1.predict(test)
correct = sum(prediction)
print("Acurracy of model is ",(correct/93)*100,"%")
```

# Dimension Reduction

To visualize our groups, we need to **reduce the dimension**(columns) by using dimension reduction technique. I am using **PCA**(**P**rincipal **C**omponent **A**nalysis) technique for dimension reduction.

I am using **`sklearn.decomposition`** for PCA.

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca.fit(spine)
pca_tran = pca.transform(spine)
```

```python
pca_tran.shape
```

Now our data is transformed in 2 columns which is applicable for further process.

```python
new_spine = pd.DataFrame(data=pca_tran,columns=['PCA1','PCA2'])
new_spine['labels'] = kmeans2.labels_
new_spine.head(5)
```

```python
plt.subplots(figsize=(15,10))
plt.scatter(x=new_spine.PCA1,y=new_spine.PCA2, c=new_spine.labels, cmap=plt.cm.jet, alpha=0.8)
plt.xlabel(s='PCA1')
plt.ylabel(s='PCA2')
plt.title(s='Plot of Transformed data done through PCA')
```

### Covariance

![covar1](https://cdncontribute.geeksforgeeks.org/wp-content/uploads/covrpopu.png)

![covar2](https://cdncontribute.geeksforgeeks.org/wp-content/uploads/Covar.png)

**covariance** is a measure of the joint variability of two random variables. If the greater values of one variable mainly correspond with the greater values of the other variable, and the same holds for the lesser values, (i.e., the variables tend to show similar behavior), the covariance is positive. 

In the opposite case, when the greater values of one variable mainly correspond to the lesser values of the other, (i.e., the variables tend to show opposite behavior), the covariance is negative. 

The sign of the covariance therefore shows the tendency in the linear relationship between the variables. The magnitude of the covariance is not easy to interpret because it is not normalized and hence depends on the magnitudes of the variables. The normalized version of the covariance, the correlation coefficient, however, shows by its magnitude the strength of the linear relation.

### Principal Component Analysis

![PCA1](https://cdn-images-1.medium.com/max/2000/1*WRKdN-NYF0mMumhfOXVa2Q.png)

**Principal Component Analysis** (PCA) is a **dimension-reduction** tool that can be used to **reduce** a large set of variables to a small set that still contains most of the information in the large set.

```python
print ('Explained variance by component: %s' % pca.explained_variance_ratio_)
print (pd.DataFrame(pca.components_,columns=spine.columns))
```

Above we can see that **pelvic_incidence**, **pelvic_tilt**, **lumbar_lordosis_angle**, **sacral_slope** and **degree_spondylolisthesis** are the variables having good influence on transformed data.

Therefore to identify our group we will use these variables. 

Now i am dividing the data according to labels and ploting correlation plot, only using above above variables. 

```python
spine_new = spine.drop(['pelvic_radius','pelvic_slope','Direct_tilt','thoracic_slope','cervical_tilt','sacrum_angle','scoliosis_slope'], axis=1)
spine_new['labels'] = kmeans2.labels_
```

```python
spine_new.head(5)
```

```python
group0 = spine_new[spine_new['labels'] == 0]
group1 = spine_new[spine_new['labels'] == 1]
print(group0.shape)
print(group1.shape)
```

```python
plt.figure(0)
f, ax = plt.subplots(figsize=(10, 5))
corr = group0.corr()
hm = sns.heatmap(round(corr,2), annot=True, ax=ax, cmap="coolwarm",fmt='.2f',
                 linewidths=.05)
f.subplots_adjust(top=0.93)
t= f.suptitle('Group 0 Attributes Correlation Heatmap', fontsize=14)


plt.figure(1)
f, ax = plt.subplots(figsize=(10, 5))
corr = group1.corr()
hm = sns.heatmap(round(corr,2), annot=True, ax=ax, cmap="coolwarm",fmt='.2f',
                 linewidths=.05)
f.subplots_adjust(top=0.93)
t= f.suptitle('Group 1 Attributes Correlation Heatmap', fontsize=14)
```

Above we can see that in **group 0**, relationship is having higher magnitude than that of **group 1** (for majority of variables), threfore, using all above results i am setting patients of **group 0** as **abnormal** means **having back problem** and patients of **group 1** as **normal** means **not having back problem**.
